```python
import pandas as pd
df=pd.read_csv("/home/user/Downloads/portland_preprocessing_dataset.csv")
```

```python
import matplotlib.pyplot as plt
plt.hist(df["price"],bins=20)
plt.title("price Disribution")
plt.xlabel("price")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.scatter(df["yearBuilt"],df["price"])
plt.title("year Built vc price")
plt.xlabel("YearBuilt")
plt.ylabel("price")
plt.show()
```

```python
df_year=df.sort_values("yearBuilt")
plt.figure()
plt.fill_between(df_year["yearBuilt"],df_year["price"])
plt.title("areaplot:year built vs price")
plt.xlabel("yearubilt")
plt.ylabel("price")
plt.show()
```

```python
df["homeType_APARTMENT"].value_counts().plot(kind="bar")
plt.title("home type distribution")
plt.xlabel("HomeType")
plt.ylabel("Count")
plt.show()
```

```python
sample=df[["bedrooms","bathrooms"]].head(100)
plt.figure()
plt.stackplot(range(len(sample)),sample["bedrooms"],sample["bathrooms"])
plt.title("STACKED AREA PLOT:BEDROOMS AND BATHROOMS")
plt.xlabel("house index")
plt.ylabel("normalized values")
plt.show()
```

```python

bedroom_counts = df['bedrooms'].value_counts()
plt.pie(bedroom_counts, labels=bedroom_counts.index, autopct='%1.1f%%')
plt.title("Bedrooms Distribution")
plt.show()
```

```python
home_counts = df['homeType_APARTMENT'].value_counts()
plt.pie(home_counts, labels=home_counts.index, autopct='%1.1f%%')
plt.title("Home Type Distribution")
plt.show()
```

```python
bath_counts = df['bathrooms'].value_counts()
plt.pie(bath_counts, labels=bath_counts.index, autopct='%1.1f%%')
plt.title("Bathrooms Distribution")
plt.show()
```

```python
# Step 1: Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 2: Load the housing dataset
df = pd.read_csv("/home/user/Downloads/portland_preprocessing_dataset(1).csv")  # Replace with your dataset path

# Step 3: Compute correlation matrix
corr_matrix = df.corr()

# Step 4: Plot the correlation heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap of Housing Dataset")
plt.show()
```
