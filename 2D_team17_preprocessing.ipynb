```python
df_clean.loc[:,"bedrooms"]=df_clean["bedrooms"].fillna(df_clean["bedrooms"].mean())
```

```python
df_clean.loc[:,'latitude'] = df_clean['latitude'].fillna(df_clean['latitude'].median())
```

```python
df_clean.loc[:,"longitude"] = df_clean["longitude"].fillna(df_clean["longitude"].median())
```

```python
df_clean.loc[:,"livingArea"] = df_clean["livingArea"].fillna(df_clean["livingArea"].median())
```

```python
df_clean.loc[:,"lotSize"] = df_clean["lotSize"].fillna(df_clean["lotSize"].median())
```

```python
df_clean.loc[:,"yearBuilt"] = df_clean["yearBuilt"].fillna(df_clean["yearBuilt"].mean())
```

```python
df_clean.loc[:,"daysOnZillow"] = df_clean["daysOnZillow"].fillna(df_clean["daysOnZillow"].median())
```

```python
df_clean.loc[:,"propertyTaxRate"] = df_clean["propertyTaxRate"].fillna(df_clean["propertyTaxRate"].median())
```

```python
num_cols=[
    'price',
    'bedrooms',
    'bathrooms',
    'livingArea',
    'lotSize',
    'address/zipcode',
    'latitude',
    'longitude',
    'dateSold',
    'lastSoldPrice',
    'yearBuilt',
    'daysOnZillow',
    'propertyTaxRate'
]
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
df_clean[num_cols]=scaler.fit_transform(df_clean[num_cols])
```

```python
df_clean[num_cols].head()
```

```python
price_bins=[0,0.33,0.66,1.0]
price_labels=["Low","Median","High"]
df_clean["price_bin"]=pd.cut(df_clean["price"],bins=price_bins,labels=price_labels)
```

```python
area_bins=[0,0.33,0.66,1.0]
area_labels=["small","Median","large"]
df_clean["livingArea_bin"]=pd.cut(df_clean["livingArea"],bins=area_bins,labels=area_labels)
```
